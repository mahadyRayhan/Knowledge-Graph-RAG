```python
import numpy as np
import matplotlib.pyplot as plt

# Data
data_llm = ['1', '3', '5', '10']

M_G = [0.737, 0.737, 0.737, 0.751]
M_P = [0.579, 0.614, 0.642, 0.7]
S_G = [0.842, 0.842, 0.842, 0.842]
S_P = [0.579, 0.526, 0.558, 0.557]

# Bar width and positions
bar_width = 0.2
x = np.arange(len(data_llm))
plt.rcParams['font.family'] = 'Times New Roman'


plt.figure(figsize=(10, 6))  # Expanded x-axis width

# Hatch patterns
hatch_patterns = ['////', '.....', 'xxx', '\\\\\\\\']

# Plot bars
bars1 = plt.bar(x - 1.5 * bar_width, M_G, width=bar_width, label='Movie-Gemini', color='white', edgecolor='black', hatch=hatch_patterns[0])
bars2 = plt.bar(x - 0.5 * bar_width, M_P, width=bar_width, label='Movie-Gpt', color='white', edgecolor='black', hatch=hatch_patterns[1])
bars3 = plt.bar(x + 0.5 * bar_width, S_G, width=bar_width, label='Scholar-Gemini', color='white', edgecolor='black', hatch=hatch_patterns[2])
bars4 = plt.bar(x + 1.5 * bar_width, S_P, width=bar_width, label='Scholar-Gpt', color='white', edgecolor='black', hatch=hatch_patterns[3])

# Adding value labels on top of each bar
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, height + 0.005, f'{height:.3f}', 
                 ha='center', va='bottom', fontsize=15, fontweight='bold', rotation=45)

add_labels(bars1)
add_labels(bars2)
add_labels(bars3)
add_labels(bars4)

# Labels and legend
plt.xlabel('Top K', fontsize=20, fontweight='bold')
plt.ylabel('Precision', fontsize=20, fontweight='bold')
plt.xticks(ticks=x, labels=data_llm, fontsize=15, fontweight='bold')
plt.yticks(fontsize=15, fontweight='bold')
plt.legend(fontsize=15, frameon=True)
plt.title('Precision on Different Dataset & LLM', fontsize=16, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, 0.8 + 0.2)  # Extend y-axis to fit legend inside the plot

save_path = 'graphs/Precision-Different-Dataset-LLM.png'
plt.savefig(save_path, dpi=300, bbox_inches='tight')
# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Times New Roman'


# Data
metrics = ["ROUGE-1", "ROUGE-2", "ROUGE-L", "BLEU", "Cosine Similarity"]

movie_gemini = [0.646975066, 0.519968115, 0.625905679, 0.343702231, 0.748219902]
movie_gpt = [0.607131079, 0.380223279, 0.498709668, 0.251305058, 0.771268286]
scholar_gemini = [0.800846893, 0.769813038, 0.76970308, 0.721665698, 0.833788009]
scholar_gpt = [0.645634205, 0.528996084, 0.553731964, 0.387898562, 0.703625095]

x = np.arange(len(metrics))  # X-axis positions

plt.figure(figsize=(10, 6))
plt.plot(x, movie_gemini, marker='o', linestyle='-', label='Movie-Gemini')
plt.plot(x, movie_gpt, marker='s', linestyle='--', label='Movie-GPT')
plt.plot(x, scholar_gemini, marker='^', linestyle='-', label='Scholar-Gemini')
plt.plot(x, scholar_gpt, marker='d', linestyle='--', label='Scholar-GPT')

# Labels and Title
plt.xticks(ticks=x, labels=metrics, fontsize=15, fontweight='bold')
plt.xlabel('LLM Metrics', fontsize=20, fontweight='bold')
plt.ylabel('Score', fontsize=20, fontweight='bold')
plt.yticks([0.0, 0.2, 0.4, 0.6, 0.8],fontsize=20, fontweight='bold')
plt.xticks(x, metrics, rotation=25)
plt.title("Comparison of Model Performance{LLM)}", fontsize=16, fontweight='bold')
plt.legend(fontsize=15, frameon=True, loc = 'lower left')
plt.grid(linestyle='--', alpha=0.7)
plt.ylim(0, 0.8 + 0.1)  # Extend y-axis to fit legend inside the plot

save_path = 'graphs/LLM-performence.png'
plt.savefig(save_path, dpi=300, bbox_inches='tight')
# Show Plot
plt.show()
```
